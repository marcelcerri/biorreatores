```python
#importando as bibliotecas
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#Entrada das variveis para o modelo
mimax = 0.3 #unidade 1/hora - taxa especfica de crescimento
ks = 5 #unidade g/L - constante de semi-saturao
Cx0 = 0.55 # unidade g/L - concentrao inicial de microrganismo
Cs0 = 30 # unidade g/L - concentrao inicial de substrato
tf = 24 #unidade horas - tempo final da integrao
Yxs = 0.2 #unidade g clulas/g substrato - coeficiente estequiomtrico

#Funo com os balanos de massa
Concini = [Cx0,Cs0]

def crescmicrobiano (Concent, t):
    Cx, Cs = Concent
    mi = mimax*(Cs/(ks + Cs))
    dCxdt = mi * Cx
    dCsdt = - (1/Yxs)*mi*Cx
    return dCxdt,dCsdt

#Vetor tempo
t = np.linspace (0,tf,101)
#integrao numrica
C = odeint(crescmicrobiano, Concini, t)

mi = mimax*(C[:,1]/(ks + C[:,1]))

plt.plot(C[:,1], mi, linewidth=2)
plt.xlabel("Concentrao de substrato (g/L)")
plt.ylabel("mi (1/h)")
plt.show()

plt.plot(t, C[:,0], linewidth=2)
plt.plot(t, C[:,1], linewidth=2)
plt.xlabel("tempo (h)")
plt.ylabel("concentrao (g/L)")
plt.show()
```
